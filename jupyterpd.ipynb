```python
import pandas as pd
```

```python
raw_data = pd.read_csv('data.csv')
```

```python
data = raw_data.iloc[::-1]
```

```python
data.pop("Unnamed: 5")
data.pop("Unnamed: 6")
data.pop("Unnamed: 7")
data.pop("Unnamed: 8")
data.pop("Unnamed: 9")
data.pop("Unnamed: 10")
data.pop("Unnamed: 11")
```

```python
data
```

```python
clean_data = data.reset_index(drop=True)
```

```python
clean_data
```

```python
pd.options.mode.chained_assignment = None
```

```python
data_close = clean_data['Close'].str.replace(',','')
```

```python
clean_data['SMA'] = data_close.rolling(window =21).mean()
```

```python
import datetime
clean_data['Day of Week'] = [datetime.datetime.strptime(item, '%b  %d, %Y').strftime("%A") for item in clean_data['Date'] ]
```

```python
friday_closings = clean_data.loc[clean_data['Day of Week'] == "Friday"] ['Close']
fc = friday_closings.str.replace(',','').astype(float)
fcr = fc.reset_index(drop=True)
```

```python
monday_openings = clean_data.loc[clean_data['Day of Week'] == "Monday"] ['Open']
mo = monday_openings.str.replace(',','').astype(float)
mor = mo.reset_index(drop=True)
```

```python
clean_data
```

```python
clean_data['Date']
```

```python
clean_data
```

```python
clean_data['Friday Closing Date'] = clean_data.loc[clean_data['Day of Week'] == "Friday"] ['Date']
clean_data['Friday Closing'] = friday_closings
clean_data['Monday Opening Date'] = clean_data.loc[clean_data['Day of Week'] == "Monday"] ['Date']
clean_data['Monday Opening'] = monday_openings
```

```python
clean_data["Friday Closing"]
```

```python
clean_data['Raw Changes'] = fcr.sub(mor)
```

```python
import numpy
clean_data['Net Changes'] = numpy.NaN
```

```python
clean_data['Net Changes'][0] = clean_data['Raw Changes'][0]
i =1
for j in range(1, len(clean_data['Close'])):
    clean_data['Net Changes'][i+4] = clean_data['Raw Changes'][j]
    i+=4
    
    
```

```python
clean_data['Net percentage of Changes'] = numpy.NaN
```

```python
perc_changes = clean_data['Raw Changes']/fcr
clean_data['raw percentage of changes'] = perc_changes*100
```

```python
clean_data['Net percentage of Changes'][0] = clean_data['raw percentage of changes'][0]
i =1
for j in range(1, len(clean_data['Close'])):
    clean_data['Net percentage of Changes'][i+4] = clean_data['raw percentage of changes'][j]
    i+=4
    
```

```python
clean_data.head(21)
```

```python
clean_data = clean_data[['Date','Close','Open','High','Low','SMA','Day of Week', 'Net Changes', 'Net percentage of Changes']]
```

```python
clean_data.to_csv('updated.csv')
```
